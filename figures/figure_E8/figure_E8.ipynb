```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
# load the source data for each panel 
df = pd.read_excel('./figure_E8_sourcedata.xlsx',engine='openpyxl',sheet_name='Panel_a',index_col=0)

eng_hga = df['eng_hga']
eng_lfs = df['eng_lfs']
spa_hga = df['spa_hga']
spa_lfs = df['spa_lfs']
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

fig,axs = plt.subplots(1,2,figsize=(10,4))

ax = axs[0]
mpl.rcParams['axes.spines.top'] = False
mpl.rcParams['axes.spines.right'] = False
mpl.rcParams['font.sans-serif'] = ['Arial']
plt.rcParams.update({'font.size': 10})#, 'font.sans-serif': 'Arial'})
ax.scatter(eng_hga,eng_lfs,alpha=0.8,clip_on=False)
ax.axhline(np.percentile(eng_lfs,90),color='k',linestyle='--')
ax.axvline(np.percentile(eng_hga,90),color='k',linestyle='--')

ax.set_ylabel('English HGA',labelpad=-7)
ax.set_xlabel('English LFS',labelpad=-7)
ax.set_xticks([0,1])
ax.set_yticks([0,1])
ax.set_ylim([0,1])
ax.set_xlim([0,1])
ax.set_title('English electrode contributions')
sns.despine(ax=ax, offset=dict(left=5, bottom=5))

ax = axs[1]
ax.scatter(spa_hga,spa_lfs,alpha=0.8,clip_on=False)
ax.axhline(np.percentile(spa_lfs,90),color='k',linestyle='--')
ax.axvline(np.percentile(spa_hga,90),color='k',linestyle='--')

ax.set_ylabel('Spanish HGA',labelpad=-7)
ax.set_xlabel('Spanish LFS',labelpad=-7)
ax.set_xticks([0,1])
ax.set_yticks([0,1])
ax.set_ylim([0,1])
ax.set_xlim([0,1])
ax.set_title('Spanish electrode contributions')
sns.despine(ax=ax, offset=dict(left=5, bottom=5))


plt.savefig('./figure_E8.jpg',dpi=300)
```
