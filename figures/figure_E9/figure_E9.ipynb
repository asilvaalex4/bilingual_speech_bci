```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
# load the source data for each panel 
df = pd.read_excel('./figure_E9_sourcedata.xlsx',engine='openpyxl',sheet_name='Panel_a',index_col=0)
all_labs_ = df['all_labs'].values
all_preds_ = df['all_preds'].values
utts = pd.read_excel('./figure_E9_sourcedata.xlsx',engine='openpyxl',sheet_name='Panel_b',index_col=0).values.squeeze()
```

```python
plt.rcParams.update({'font.size': 10})
plt.rcParams.update({'font.sans-serif':'Arial'})

conf = confusion_matrix(all_labs_, all_preds_)
plt.figure(dpi=150,figsize=(20,20))

im = plt.imshow(conf/np.sum(conf,axis=1),cmap='binary',vmin=0,vmax=1)
plt.xticks(range(0,np.max(all_labs_)+1),utts,rotation=90)
plt.yticks(range(0,np.max(all_labs_)+1),utts,rotation=0)
plt.colorbar(im,fraction=0.05,label='Confusion',aspect=30)
plt.savefig('./figure_E9.jpg',dpi=300,bbox_inches='tight')
plt.show()
```
