```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
# load the source data for each panel 
df = pd.read_excel('./SD_ED_FIG4.xlsx',engine='openpyxl',sheet_name='Panel_a',index_col=0)
```

```python
import matplotlib as mpl
mpl.rcParams['axes.spines.top'] = False
mpl.rcParams['axes.spines.right'] = False
mpl.rcParams['font.sans-serif'] = ['Arial']
plt.rcParams.update({'font.size': 10})
fig,ax = plt.subplots(figsize=(5,4))
sns.boxplot(data=df,x='Paradigm',y='Accuracy',color=sns.color_palette().as_hex()[0])
ax.axhline(10,linestyle='--',color='k')
ax.set_ylabel('Accuracy of model trained on attempted speech (%)')
ax.set_xlabel('Evaluation paradigm')
ax.set_yticks([0,15,30,45,60,75,90])
ax.set_ylim([0,90])
sns.despine(ax=ax, offset=dict(left=5, bottom=5))
plt.savefig('./ED_FIG4.jpg',dpi=300,bbox_inches='tight')
```
