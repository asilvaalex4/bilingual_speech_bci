```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
# load the source data for each panel 
df = pd.read_excel('./SD_ED_FIG6.xlsx',engine='openpyxl',sheet_name='Panel_a',index_col=0)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
from statannot import add_stat_annotation
colors = ['#b20d30','#c17817','#3f84e5','#3f784c']

fig,ax = plt.subplots(figsize=(5,4))
mpl.rcParams['axes.spines.top'] = False
mpl.rcParams['axes.spines.right'] = False
mpl.rcParams['font.sans-serif'] = ['Arial']
plt.rcParams.update({'font.size': 10})#, 'font.sans-serif': 'Arial'})

sns.boxplot(ax=ax,data=df,x='Language',y='Mean pairwise MCD',palette=colors)
test_results = add_stat_annotation(ax, data=df, x='Language',y='Mean pairwise MCD', box_pairs = [('English','Spanish')],
                                   test='Mann-Whitney', text_format='star',
                                   loc='inside', verbose=2,line_offset=0.01)
ax.set_ylabel('Mean pairwise mel-cepstral distortion (MCD)')

ax.set_yticks([5.5,6,6.5,7,7.5,8,8.5,9,9.5])
ax.set_ylim([5.5,9.5])
sns.despine(ax=ax, offset=dict(left=5, bottom=5))

plt.savefig('./ED_FIG6.jpg',dpi=300,bbox_inches='tight')
```
